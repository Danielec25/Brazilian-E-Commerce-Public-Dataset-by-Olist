```python
#importando a biblioteca
import pandas as pd
import plotly.express as px

#visualização da base
tabela_sellers = pd.read_csv("olist_sellers_dataset.csv", sep=",")

#exibição da tabela
display(tabela_sellers)


# Excluindo mais de uma coluna
colunas_para_excluir = ["seller_id"]
tabela_sellers = tabela_sellers.drop(columns=colunas_para_excluir)
display(tabela_sellers)


#excluir linhas vazias
tabela_sellers = tabela_sellers.dropna()

#vizualizar as informações da tabela
display(tabela_sellers.info())

# Ordenando os dados por uma coluna específica (por exemplo, 'category_name')
tabela_ordenada = tabela_sellers.sort_values(by='seller_state')
display(tabela_ordenada)


# Gerando Gráficos com comparações
for coluna in tabela_sellers.columns:
    #Criando os gráficos da tabela
    grafico = px.histogram(tabela_sellers, x=coluna)
  #Exibir o gráfico
    grafico.show()
```
