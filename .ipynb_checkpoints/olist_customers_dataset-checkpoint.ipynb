```python
import pandas as pd

tabela = pd.read_csv("olist_customers_dataset.csv", sep=",")
display(tabela)

# Visualizando os primeiros registros
print("Primeiros registros da tabela:")
print(tabela.head())

# Ordenando os dados por uma coluna específica (por exemplo, 'category_name')
tabela_ordenada = tabela.sort_values(by='customer_city')

# Visualizando os primeiros registros após a ordenação
print("\nPrimeiros registros da tabela após ordenação:")
print(tabela_ordenada.head())

# Filtrando os dados (por exemplo, onde 'category_name' é igual a algum valor específico)
tabela_filtrada = tabela[tabela['customer_city'] == 'Cidade do Cliente']

# Visualizando os primeiros registros após a filtragem
print("\nPrimeiros registros da tabela após filtragem:")
print(tabela_filtrada.head())
```
