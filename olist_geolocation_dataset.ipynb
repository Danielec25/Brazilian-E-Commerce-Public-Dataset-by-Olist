```python
#importando a biblioteca
import pandas as pd
import plotly.express as px


#visualização da base
tabela = pd.read_csv("olist_geolocation_dataset.csv", sep=",")

#exibição da tabela
display(tabela)

# Ordenando os dados por uma coluna específica (por exemplo, 'category_name')
tabela_ordenada = tabela.sort_values(by='geolocation_zip_code_prefix')
display(tabela_ordenada)


#excluir linhas vazias
tabela = tabela.dropna()

#vizualizar as informações da tabela
display(tabela.info())

# Gerando Gráficos com comparações

for coluna in tabela.columns:
    #Criando os gráficos da tabela
    grafico = px.histogram(tabela, x=coluna)
  #Exibir o gráfico
    grafico.show()

```
