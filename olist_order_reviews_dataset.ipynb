```python
#importando a biblioteca
import pandas as pd
import plotly.express as px

#visualização da base
tabela_order_reviews = pd.read_csv("olist_order_reviews_dataset.csv", sep=",")

#exibição da tabela
display(tabela_order_reviews)

#excluir linhas vazias
tabela_order_reviews = tabela_order_reviews.dropna()

#vizualizar as informações da tabela
display(tabela_order_reviews.info())

# Ordenando os dados por uma coluna específica (por exemplo, 'category_name')
tabela_ordenada = tabela_order_reviews.sort_values(by='review_score')
display(tabela_ordenada)

# Gerando Gráficos com comparações
for coluna in tabela_order_reviews.columns:
    #Criando os gráficos da tabela
    grafico = px.histogram(tabela_order_reviews, x=coluna)
  #Exibir o gráfico
    grafico.show()


# Convertendo as colunas de data para o formato datetime
tabela_order_reviews['review_creation_date'] = pd.to_datetime(tabela_order_reviews['review_creation_date'])
tabela_order_reviews['review_answer_timestamp'] = pd.to_datetime(tabela_order_reviews['review_answer_timestamp'])

# Calculando o tempo de resposta
tabela_order_reviews['tempo_resposta'] = tabela_order_reviews['review_answer_timestamp'] - tabela_order_reviews['review_creation_date']

# Exibindo a tabela com o tempo de resposta
display(tabela_order_reviews[['review_score', 'tempo_resposta']])

# Criando um gráfico de dispersão para visualizar a relação entre pontuação e tempo de resposta
grafico_tempo_resposta = px.scatter(tabela_order_reviews, x='tempo_resposta', y='review_score', title='Relação entre Pontuação e Tempo de Resposta')
grafico_tempo_resposta.show()
```
